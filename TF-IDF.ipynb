```python
import json
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
%cd drive/MyDrive/cs182_project/
%load_ext autoreload
%autoreload 2
```

```python
with open('yelp_review_training_dataset.jsonl', 'r') as json_file:
    json_list = list(json_file)
```

```python
reviews = {}
for json_str in json_list:
    result = json.loads(json_str)
    vectorizer = TfidfVectorizer(max_features=50)
    try:
      X = vectorizer.fit_transform([result['text']])
    except: 
      continue 
    reviews[result['text']] = vectorizer.get_feature_names()
```

```python
print(reviews)
```

```python
with open("tf-idf.json", "w") as write_file:
    json.dump(reviews, write_file, indent=4)
write_file.close()
```
